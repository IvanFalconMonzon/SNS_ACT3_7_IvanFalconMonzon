# **Paso 4: Evaluar los modelos usando Lazy Predict**

Lazy Predict permite evaluar rápidamente diferentes modelos.

```python
# IVAN FALCON MONZON
!pip install lazypredict --quiet  # Instalar Lazy Predict si no está instalado

# ========================
#  Paso 0: Importar Librerías necesarias
# ========================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# ========================
#  Paso 1: Cargar datos
# ========================
train_data = pd.read_csv('train_values.csv')
train_labels = pd.read_csv('train_labels.csv')

# Fusionar etiquetas con los datos de entrenamiento
train = train_data.merge(train_labels, on="building_id")

# =============================
#  Paso 2: Preprocesamiento
# =============================
# Convertir variables categóricas en numéricas
categorical_cols = train.select_dtypes(include=['object']).columns
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    label_encoders[col] = le  # Guardamos los encoders

# Definir variables X (features) e y (target)
X = train.drop(columns=["building_id", "damage_grade"])
y = train["damage_grade"]

# =============================
#  Paso 3: Reducción de Memoria
# =============================
def reduce_memory_usage(df):
    for col in df.columns:
        if df[col].dtype == "int64":
            df[col] = df[col].astype("int16")
        elif df[col].dtype == "float64":
            df[col] = df[col].astype("float16")
    return df

X_reduced = reduce_memory_usage(X)

# =================================
#  Paso 4: División de datos
# =================================
# Usamos solo el 1% del dataset para evitar problemas de RAM
X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(
    X_reduced, y, test_size=0.99, random_state=42
)

# ==============================
#  Paso 5: Lazy Predict - Solo Decision Tree y Random Forest
# ==============================
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

# Ejecutar Lazy Predict con todos los modelos disponibles
models, predictions = clf.fit(X_train_sub, X_test_sub, y_train_sub, y_test_sub)

# Mostrar los resultados si existen
if not models.empty:
    print("\nResultados de Lazy Predict:")
    display(models)
else:
    print("Error: No se generaron modelos, reducir el dataset.")
```

## **5.1: Random Forest Classifier, optimizado con RandomizedSearchCV**

- Algoritmo: Random Forest (Bosque Aleatorio)

- Optimización: RandomizedSearchCV (búsqueda aleatoria de hiperparámetros)

- Métrica usada: f1_micro (para clasificación multiclase)

- Mejor modelo: Se entrena de nuevo con los mejores hiperparámetros encontrados

- Predicción: Se generan predicciones en X_test y se guardan en un CSV

**¿Qué tipo de modelo es?**

- Clasificador basado en árboles de decisión (Random Forest)

- Modelo de ensamblado (ensemble) con múltiples árboles

- Optimizado con búsqueda aleatoria de hiperparámetros (RandomizedSearchCV)

- Métricas de evaluación: F1-score micro (para clasificación multiclase)

```python
# IVAN FALCON MONZON
# Librerías necesarias:
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.metrics import f1_score

# Definición del modelo:
rf_model = RandomForestClassifier(random_state=42)

# Definición del espacio de búsqueda para hiperparámetros:
param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20],
    'min_samples_split': [5, 10]
}

# Optimización con RandomizedSearchCV:
random_search_rf = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid_rf,
    scoring='f1_micro',
    n_iter=5,  # Número de combinaciones a probar
    cv=3,  # Usar 3 pliegues de validación cruzada
    verbose=3,
    n_jobs=-1,  # Usa todos los núcleos disponibles
    random_state=42
)

# Ajustar el modelo con RandomizedSearchCV
random_search_rf.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros y la mejor puntuación F1
print("Mejores hiperparámetros para Random Forest:", random_search_rf.best_params_)
print("Mejor puntuación F1:", random_search_rf.best_score_)

# Entrenar con los mejores hiperparámetros encontrados:
best_rf_model = RandomForestClassifier(
    n_estimators=random_search_rf.best_params_['n_estimators'],
    max_depth=random_search_rf.best_params_['max_depth'],
    min_samples_split=random_search_rf.best_params_['min_samples_split'],
    random_state=42
)

# Entrenar el modelo con los datos de entrenamiento
best_rf_model.fit(X_train, y_train)

# Evaluar el modelo en el conjunto de validación
y_pred = best_rf_model.predict(X_val)
f1 = f1_score(y_val, y_pred, average="micro")
print("F1-score en validación:", f1)

# Alinear las columnas del conjunto de test con el conjunto de entrenamiento
X_test = X_test[X_train.columns]

# Realizar las predicciones en el conjunto de test
final_predictions = best_rf_model.predict(X_test)

# Crear el DataFrame de submission
submission = pd.DataFrame({
    "building_id": test_data["building_id"],  # Asegúrate de que 'test_data' tiene la columna 'building_id'
    "damage_grade": final_predictions
})

# Guardar el archivo CSV de submission
submission.to_csv("submission.csv", index=False)
print("Archivo de submission generado con éxito.")
```

## 5.2: DecisionTreeClassifier con optimización RandomizedSearchCV

Este código implementa un modelo de clasificación basado en un árbol de decisión para predecir el grado de daño de los edificios.

Se realiza una optimización de hiperparámetros con RandomizedSearchCV, utilizando validación cruzada de 5 pliegues y búsqueda aleatoria en 10 combinaciones.

Luego, el mejor modelo encontrado se entrena con los datos de entrenamiento y se evalúa con la métrica F1-score en el conjunto de validación. Finalmente, se generan predicciones sobre el conjunto de prueba y se guardan en un archivo CSV para su posterior uso.



```python
# IVAN FALCON MONZON
# Importación de librerías necesarias
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Cargar los datos de prueba
test_data = pd.read_csv('test_values.csv')

# Definir el modelo de árbol de decisión
dt_model = DecisionTreeClassifier(random_state=42)

# Espacio de búsqueda para la optimización de hiperparámetros
param_grid_dt = {
    'max_depth': [10, 20, 30, 40, 50, None],  # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10, 20, 50],  # Mínimo de muestras para dividir un nodo
    'min_samples_leaf': [1, 2, 4, 6, 10],  # Mínimo de muestras en las hojas
    'criterion': ['gini', 'entropy']  # Criterios de división
}

# Optimización de hiperparámetros con búsqueda aleatoria y validación cruzada
random_search_dt = RandomizedSearchCV(
    estimator=dt_model,
    param_distributions=param_grid_dt,
    scoring='f1_micro',  # Métrica de evaluación F1-score micro
    n_iter=10,  # Número de combinaciones de hiperparámetros a evaluar
    cv=5,  # Validación cruzada con 5 pliegues
    verbose=3,
    n_jobs=-1,  # Paralelización usando todos los núcleos disponibles
    random_state=42
)

# Ajustar el modelo con los datos de entrenamiento
random_search_dt.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros y la mejor puntuación F1 obtenida
print("Mejores hiperparámetros para Decision Tree:", random_search_dt.best_params_)
print("Mejor puntuación F1:", random_search_dt.best_score_)

# Definir el modelo con los mejores hiperparámetros encontrados
best_dt_model = DecisionTreeClassifier(
    max_depth=random_search_dt.best_params_['max_depth'],
    min_samples_split=random_search_dt.best_params_['min_samples_split'],
    min_samples_leaf=random_search_dt.best_params_['min_samples_leaf'],
    criterion=random_search_dt.best_params_['criterion'],
    random_state=42
)

# Entrenar el modelo optimizado con los datos de entrenamiento
best_dt_model.fit(X_train, y_train)

# Evaluar el modelo en el conjunto de validación
y_pred = best_dt_model.predict(X_val)
f1 = f1_score(y_val, y_pred, average="micro")
print("F1-score en validación:", f1)

# Generar predicciones en el conjunto de prueba
final_predictions = best_dt_model.predict(X_test)

# Crear un DataFrame con los resultados de las predicciones
submission = pd.DataFrame({
    "building_id": test_data["building_id"],
    "damage_grade": final_predictions
})

# Guardar las predicciones en un archivo CSV para la competencia
submission.to_csv("submission.csv", index=False)
print("Archivo de submission generado con éxito.")
```

## 5.3: Gradient Boosting optimizando sus hiperparámetros con RandomizedSearchCV

Este código implementa un modelo de clasificación basado en Gradient Boosting, optimizando sus hiperparámetros con RandomizedSearchCV.

Se reduce la carga computacional limitando el número de árboles y la profundidad, además de aplicar validación cruzada con 3 pliegues. Luego, el mejor modelo se entrena con los datos de entrenamiento y se evalúa con la métrica F1-score en el conjunto de validación. Se incluye early stopping para detener el entrenamiento si no hay mejora tras 5 iteraciones.



```python
# IVAN FALCON MONZON
# Librerías necesarías
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import f1_score

# Definir el modelo base de Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)

# Espacio de búsqueda de hiperparámetros ajustado para eficiencia
param_grid_gb = {
    'n_estimators': [50, 100],  # Número de árboles en el ensamble
    'learning_rate': [0.05, 0.1],  # Tasa de aprendizaje para control de ajuste
    'max_depth': [3, 5],  # Profundidad máxima del árbol
    'min_samples_split': [2, 5],  # Mínimo de muestras para dividir un nodo
    'min_samples_leaf': [1, 2],  # Mínimo de muestras por hoja
    'subsample': [0.8, 1.0],  # Fracción de datos utilizada en cada iteración
    'max_features': ['sqrt']  # Selección aleatoria de características en cada división
}

# Optimización de hiperparámetros con búsqueda aleatoria y validación cruzada
random_search_gb = RandomizedSearchCV(
    estimator=gb_model,
    param_distributions=param_grid_gb,
    scoring='f1_micro',  # Métrica de evaluación F1-score micro
    n_iter=6,  # Número limitado de combinaciones para reducir el tiempo de cómputo
    cv=3,  # Validación cruzada con 3 pliegues para menor carga computacional
    verbose=3,
    n_jobs=-1,  # Paralelización en todos los núcleos disponibles
    random_state=42
)

# Ajustar el modelo con los datos de entrenamiento
random_search_gb.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros y la mejor puntuación F1 obtenida
print("Mejores hiperparámetros para Gradient Boosting:", random_search_gb.best_params_)
print("Mejor puntuación F1 para Gradient Boosting:", random_search_gb.best_score_)

# Definir el modelo con los mejores hiperparámetros encontrados
best_gb_model = GradientBoostingClassifier(
    **random_search_gb.best_params_,
    random_state=42,
    n_iter_no_change=5  # Early stopping: se detiene si no mejora en 5 iteraciones
)

# Entrenar el modelo optimizado con los datos de entrenamiento
best_gb_model.fit(X_train, y_train)

# Evaluar el modelo en el conjunto de validación
y_pred_gb = best_gb_model.predict(X_val)
f1_gb = f1_score(y_val, y_pred_gb, average="micro")
print("F1-score en validación (Gradient Boosting):", f1_gb)
```

## 5.4: modelo de clasificación utilizando XGBoost, optimizando sus hiperparámetros con RandomizedSearchCV

Este código implementa un modelo de clasificación utilizando XGBoost, optimizando sus hiperparámetros con RandomizedSearchCV.

Se realiza un ajuste en las etiquetas de la variable objetivo (y_train_adj, y_val_adj) para que sean compatibles con el modelo.

Tras la optimización, el mejor modelo se entrena con los datos de entrenamiento y se evalúa utilizando F1-score en el conjunto de validación. Finalmente, se generan predicciones para el conjunto de prueba y se guardan en un archivo CSV.



```python
# IVAN FALCON MONZON
# Librerías necesarías
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import f1_score
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Cargar datos de prueba
test_data = pd.read_csv('test_values.csv')  # Asegurar que el archivo de test sea el correcto

# Ajustar etiquetas de la variable objetivo si es necesario
y_train_adj = y_train - 1  # Ajuste de etiquetas para alinearlas con XGBoost
y_val_adj = y_val - 1

# Definir el modelo base de XGBoost
xgb_model = XGBClassifier(
    objective="multi:softmax",  # Clasificación multiclase con etiquetas discretas
    eval_metric="mlogloss",  # Función de evaluación: log-loss multiclase
    use_label_encoder=False,  # Desactivar el codificador de etiquetas de XGBoost
    random_state=42
)

# Espacio de búsqueda de hiperparámetros
param_grid_xgb = {
    'n_estimators': [100, 200],  # Número de árboles en el ensamble
    'learning_rate': [0.05, 0.1, 0.2],  # Tasa de aprendizaje para el ajuste de pesos
    'max_depth': [3, 5, 7],  # Profundidad máxima de los árboles
    'min_child_weight': [1, 3, 5],  # Peso mínimo requerido en un nodo hoja
    'gamma': [0, 0.1, 0.2],  # Reducción mínima de pérdida para realizar un split
    'subsample': [0.8, 1.0],  # Proporción de datos utilizados por cada árbol
    'colsample_bytree': [0.8, 1.0]  # Proporción de características consideradas en cada split
}

# Optimización de hiperparámetros con RandomizedSearchCV
random_search_xgb = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid_xgb,
    scoring='f1_micro',  # Evaluación con F1-score micro
    n_iter=6,  # Número de combinaciones de hiperparámetros a probar
    cv=3,  # Validación cruzada con 3 pliegues para reducir carga computacional
    verbose=3,
    n_jobs=-1,  # Paralelización en todos los núcleos disponibles
    random_state=42
)

# Ajustar el modelo con los datos de entrenamiento y etiquetas ajustadas
random_search_xgb.fit(X_train, y_train_adj)

# Imprimir los mejores hiperparámetros encontrados y la mejor puntuación F1 obtenida
print("Mejores hiperparámetros para XGBoost:", random_search_xgb.best_params_)
print("Mejor puntuación F1 para XGBoost:", random_search_xgb.best_score_)

# Definir el modelo final con los mejores hiperparámetros encontrados
best_xgb_model = XGBClassifier(
    **random_search_xgb.best_params_,
    objective="multi:softmax",
    eval_metric="mlogloss",
    use_label_encoder=False,
    random_state=42
)

# Ajustar el modelo final con los datos de entrenamiento
best_xgb_model.fit(X_train, y_train_adj)

# Evaluar el modelo en el conjunto de validación
y_pred_xgb = best_xgb_model.predict(X_val)
f1_xgb = f1_score(y_val_adj, y_pred_xgb, average="micro")
print("F1-score en validación (XGBoost):", f1_xgb)

# Realizar predicciones sobre el conjunto de test
y_test_pred_xgb = best_xgb_model.predict(X_test)

# Crear el DataFrame para la entrega de resultados
submission = pd.DataFrame({
    "building_id": test_data["building_id"],
    "damage_grade": y_test_pred_xgb + 1  # Ajustar etiquetas si es necesario
})

# Guardar las predicciones en un archivo CSV
submission.to_csv("submission.csv", index=False)
print("Archivo de submission generado con éxito.")
```

## 5.5: SVM con RandomizedSearchCV.

Este código entrena un modelo SVM (Support Vector Machine) con LinearSVC, optimizando la regularización con RandomizedSearchCV.

Se aplica escalado estándar (StandardScaler) para mejorar la estabilidad del modelo, y se usa una muestra reducida del conjunto de entrenamiento para acelerar el proceso.

Finalmente, se generan predicciones y se guardan en un archivo CSV.

**Puntos clave**

- Muestreo: Se seleccionan 5000 muestras aleatorias del conjunto de entrenamiento (train_sample) para optimizar el modelo más rápido.
- Escalado: Se normalizan los datos con StandardScaler(), lo cual es fundamental para que SVM funcione correctamente.
- Optimización: Se ajusta el hiperparámetro C mediante RandomizedSearchCV con solo 3 iteraciones y 2 folds para reducir la carga computacional.
- Evaluación: Se mide el F1-score (micro) en el conjunto de validación.
- Predicciones finales: Se generan predicciones sobre el conjunto de test y se guardan en submission.csv.


```python
# IVAN FALCON MONZON
# Librerías necesarías
from sklearn.svm import LinearSVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
import pandas as pd
import numpy as np

# Definir tamaño de muestra para entrenar más rápido
TEST_SAMPLE_SIZE = 5000

# Seleccionar una muestra aleatoria del conjunto de entrenamiento
train_sample = X_train.sample(TEST_SAMPLE_SIZE, random_state=42)
train_labels = y_train.loc[train_sample.index]

# Escalar los datos con StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_sample)  # Ajustar y transformar datos de entrenamiento
X_val_scaled = scaler.transform(X_val)  # Transformar datos de validación
X_test_scaled = scaler.transform(X_test)  # Transformar datos de prueba

# Definir el modelo SVM con LinearSVC
svm_model = LinearSVC(random_state=42, dual=False, max_iter=5000)

# Espacio de búsqueda de hiperparámetros para SVM
param_grid_svm = {
    'C': [0.1, 1, 10]  # Parámetro de regularización
}

# Optimización con RandomizedSearchCV
random_search_svm = RandomizedSearchCV(
    estimator=svm_model,
    param_distributions=param_grid_svm,
    scoring='f1_micro',  # Optimización basada en F1-score micro
    n_iter=3,  # Reducimos las iteraciones para mayor velocidad
    cv=2,  # Usamos solo 2 folds en validación cruzada para reducir tiempo
    verbose=1,
    n_jobs=-1,  # Utilizar todos los núcleos disponibles
    random_state=42
)

# Entrenar el modelo con la muestra de datos escalados
random_search_svm.fit(X_train_scaled, train_labels)

# Mostrar los mejores hiperparámetros encontrados
print("Mejores hiperparámetros para SVM:", random_search_svm.best_params_)
print("Mejor puntuación F1 para SVM:", random_search_svm.best_score_)

# Modelo final con los mejores hiperparámetros encontrados
best_svm_model = LinearSVC(
    C=random_search_svm.best_params_['C'],
    random_state=42,
    dual=False,
    max_iter=5000
)

# Entrenar el modelo final con los datos escalados
best_svm_model.fit(X_train_scaled, train_labels)

# Evaluar en el conjunto de validación
y_pred_svm = best_svm_model.predict(X_val_scaled)
f1_svm = f1_score(y_val, y_pred_svm, average="micro")
print("F1-score en validación (SVM):", f1_svm)

# Realizar predicciones sobre el conjunto de test
final_predictions = best_svm_model.predict(X_test_scaled)

# Crear archivo de submission
submission = pd.DataFrame({
    "building_id": test_data["building_id"],
    "damage_grade": final_predictions
})
submission.to_csv("submission.csv", index=False)
print("Archivo de submission generado con éxito.")
```
